Programming Assignment 1: matrix factorization using gradient descent 

```python
import numpy as np
import pandas as pd
import progressbar as pb
import numpy
```

```python
def Matrix_Factorization(T1, P1, P2, T2, steps=1000, alpha=0.0002, beta=0.02):
    P2 = P2.T #Transpose of matrix P2
    for step in pb.progressbar(range(steps)): # to keep track of loops
        for i in range(len(T1)):
            for j in range(len(T1[i])):
                if T1[i][j] > 0:
                    eij = T1[i][j] - numpy.dot(P1[i,:],P2[:,j])
                    for k in range(T2):
                        P1[i][k] = P1[i][k] + alpha * (2 * eij * P2[k][j] - beta * P1[i][k])
                        P2[k][j] = P2[k][j] + alpha * (2 * eij * P1[i][k] - beta * P2[k][j])
        eR = numpy.dot(P1,P2)
        e = 0
        for i in range(len(T1)):
            for j in range(len(T1[i])):
                if T1[i][j] > 0:
                    e = e + pow(T1[i][j] - numpy.dot(P1[i,:],P2[:,j]), 2)
                    for k in range(T2):
                        e = e + (beta/2) * (pow(P1[i][k],2) + pow(P2[k][j],2))
        if e < 0.001:
            break
    return P1, P2.T
```
